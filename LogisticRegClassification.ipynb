```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
dataset = pd.read_csv("diabetes.csv", usecols=['Glucose','BloodPressure','Outcome'])
```

```python
dataset
```

```python
plt.scatter(dataset["Glucose"], dataset["Outcome"])
plt.xlabel("Glucose level")
plt.ylabel("Truth value")
plt.show()
```

```python
labels = ["No Diabetes", "Diabetes"]
colors = ["green", "red"]
outcome_counts = dataset["Outcome"].value_counts()
plt.pie(outcome_counts, labels=labels, colors=colors)
plt.show()
```

```python
x = dataset[dataset.columns[:-1]].values
y = dataset[dataset.columns[-1]].values
```

```python
x,y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.8, random_state=42)
```

```python
model = LogisticRegression()
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
Y_pred
```

```python
accuracy = accuracy_score(Y_test, Y_pred)
print("Accuracy:", accuracy)
```

```python
classification_report = classification_report(Y_test, Y_pred)
print("Classification Report:\n", classification_report)
```

```python

confusion_mat = confusion_matrix(Y_test, Y_pred)
print("Confusion Matrix:\n", confusion_mat)
```

```python
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_mat, annot=True, cmap="Blues")
plt.title("Confusion Matrix")
plt.xlabel("Predicted value")
plt.ylabel("Truth value")
plt.show()
```

```python
x = dataset['Glucose'].values.reshape(-1, 1)
y = dataset['Outcome'].values
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.8, random_state=42)

model = LogisticRegression()
model.fit(X_train, Y_train)

x_range = np.linspace(np.min(x), np.max(x), num=100).reshape(-1, 1)
probabilities = model.predict_proba(x_range)[:, 1]
```

```python
plt.scatter(dataset["Glucose"], dataset["Outcome"], color='blue', label='Data')
plt.plot(x_range, probabilities, color='red', label='Logistic Regression')
plt.xlabel("Glucose level")
plt.ylabel("Probability")
plt.title("Logistic Regression Curve")
plt.legend()
plt.show()
```
